```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
#import csv
county_median = Path("../Desktop/estimate_of_median.csv")

# Read Data File and store into Pandas DataFrames
income_by_county = pd.read_csv("estimate_of_median.csv", header=1)


income_by_county.head()
```

```python
# clean columns

national_income = ['Region Name', '2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01'] 
income_by_county = income_by_county[national_income]

income_by_county
```

```python
#create new csv

national_median_yearly_income = pd.DataFrame(income_by_county)

# Write DataFrame to CSV file
national_median_yearly_income.to_csv('national_median_yearly_income.csv', index=False)
```

```python
# Define the median for
median_columns = income_by_county[['2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01']].median().round(2)

# Add the mean as a new column in the DataFrame
income_by_county['Median_Columns'] = median_columns

# Print the DataFrame
median_columns
```

```python
years = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
average_income = [41255.5, 41974.5, 43018.5, 43990.5, 45229.0, 46807.0, 47592.0, 48893.0, 50576.0, 53344.0, 55150.0]

# Plot the average income for each year
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(years, average_income, marker='o', color='b', linestyle='-')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Median National Income')
plt.title('Median National Income for Each Year')

# Add grid
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

```python
#Summary Stats
years_stats = income_by_county[['2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01']]
mean_years = years_stats.mean()
median_years = years_stats.median()
std_years = years_stats.std()
var_years = years_stats.var()
sem_years = years_stats.sem()

# Print the results
print("\nSummary Stats:")
print("Mean:\n", mean_years)
print("Median:\n", median_years)
print("Standard Deviation:\n", std_years)
print("Variance:\n", var_years)
print("Standard Error of the Mean:\n", sem_years)
```
