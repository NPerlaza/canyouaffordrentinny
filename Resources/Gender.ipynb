```python
import requests
import pandas as pd
import configparser

# Read the configuration file
config = configparser.ConfigParser()
config.read('/Users/NP/Documents/canyouaffordrentinny/Resources/config.ini')

# Get the API key from the configuration file
api_key = config['DEFAULT']['API_KEY']

# Base URL for the ACS5 API
base_url = 'https://api.census.gov/data/{year}/acs/acs5'

# FIPS codes for NYC Boroughs
borough_fips = {
    'Manhattan': '061',
    'Bronx': '005',
    'Brooklyn': '047',
    'Queens': '081',
    'Staten Island': '085'
}

# Variables to fetch - Total male population (B01001_002E) and total female population (B01001_026E)
variables = 'B01001_002E,B01001_026E'

# Create an empty DataFrame to store results
results_df = pd.DataFrame(columns=['Year', 'Borough', 'Male Population', 'Female Population', 'Male Percentage', 'Female Percentage'])

# Loop over each year from 2010 to 2020
for year in range(2010, 2021):
    # Loop over each borough
    for borough_name, fips_code in borough_fips.items():
        # Construct the request URL
        request_url = f'{base_url.format(year=year)}?get=NAME,{variables}&for=county:{fips_code}&in=state:36&key={api_key}'
        
        # Make the request
        response = requests.get(request_url)
        
        try:
            # Parse the response as JSON
            data = response.json()
            
            # Skip the header row and extract the data
            if data and len(data) > 1:
                data_row = data[1]
                total_population = int(data_row[1]) + int(data_row[2])
                
                # Create a new DataFrame row
                new_row = pd.DataFrame([{
                    'Year': year,
                    'Borough': borough_name,
                    'Male Population': int(data_row[1]),
                    'Female Population': int(data_row[2]),
                    'Male Percentage': (int(data_row[1]) / total_population) * 100,
                    'Female Percentage': (int(data_row[2]) / total_population) * 100
                }])
                
                # Append the new row to the results DataFrame
                results_df = pd.concat([results_df, new_row], ignore_index=True)
        except ValueError:
            print("Decoding JSON has failed")

# Display the DataFrame
print(results_df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert columns to correct data type
results_df['Year'] = pd.to_numeric(results_df['Year'], errors='coerce')
results_df['Male Population'] = pd.to_numeric(results_df['Male Population'], errors='coerce')
results_df['Female Population'] = pd.to_numeric(results_df['Female Population'], errors='coerce')

# Handle missing values
results_df = results_df.dropna()

# Remove duplicates
results_df = results_df.drop_duplicates()

# Separate the data for each borough
boroughs = results_df['Borough'].unique()

# Create a new figure
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart for each borough
for i, borough in enumerate(boroughs):
    borough_data = results_df[results_df['Borough'] == borough]
    ax.bar(i - 0.2, borough_data['Male Population'].sum(), 0.4, color='blue')
    ax.bar(i + 0.2, borough_data['Female Population'].sum(), 0.4, color='red')

# Add a legend
ax.legend(['Male', 'Female'])

# Add title and labels
ax.set_title('Population in NYC Boroughs')
ax.set_xlabel('Borough')
ax.set_ylabel('Population')

# Set x-ticks to be the borough names
ax.set_xticks(range(len(boroughs)))
ax.set_xticklabels(boroughs)

# Show the plot
plt.show()
```

```python
# Create a dictionary to store the population data
populations = {}

# Calculate the total male and female populations for each borough
for borough in boroughs:
    borough_data = results_df[results_df['Borough'] == borough]
    male_population = borough_data['Male Population'].sum()
    female_population = borough_data['Female Population'].sum()
    populations[borough] = {'Male': male_population, 'Female': female_population}

# Convert the dictionary to a DataFrame and print it
populations_df = pd.DataFrame(populations).T
print(populations_df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load the data into a DataFrame named 'merged_data'
merged_data = pd.read_csv('/Users/NP/canyouaffordrentinny/Resources/NYC_Gender_Population_Percentages.csv')

# Define the list of boroughs
boroughs = ['Manhattan', 'Bronx', 'Brooklyn', 'Queens', 'Staten Island']

# Choose a specific year for the pie charts
year = 2020

# Define the colors for male and female
colors = ['blue', 'orange']

# Create a figure with a subplot for each borough (one for the line chart and one for the pie chart)
fig, axs = plt.subplots(len(boroughs), 2, figsize=(12, 6*len(boroughs)))

# Loop over each borough
for i, borough in enumerate(boroughs):
    # Select data for the current borough
    borough_data = merged_data[merged_data['Borough'] == borough]

    # Calculate total male and female population for each year
    total_male_population = borough_data.groupby('Year')['Male Population'].sum()
    total_female_population = borough_data.groupby('Year')['Female Population'].sum()

    # Create a line plot on the first subplot
    axs[i, 0].plot(total_male_population.index, total_male_population, label='Male', color=colors[0])
    axs[i, 0].plot(total_female_population.index, total_female_population, label='Female', color=colors[1])

    # Add linear regression line and print model's formula and r values for male population
    X = sm.add_constant(total_male_population.index)  # adding a constant
    model = sm.OLS(total_male_population, X).fit()
    predictions = model.predict(X) 
    axs[i, 0].plot(total_male_population.index, predictions, 'r', label='Male Regression Line')
    print(f"{borough} Male Population Regression: {model.params[0]} + {model.params[1]}*Year, R-squared: {model.rsquared}")

    # Add linear regression line and print model's formula and r values for female population
    X = sm.add_constant(total_female_population.index)  # adding a constant
    model = sm.OLS(total_female_population, X).fit()
    predictions = model.predict(X) 
    axs[i, 0].plot(total_female_population.index, predictions, 'g', label='Female Regression Line')
    print(f"{borough} Female Population Regression: {model.params[0]} + {model.params[1]}*Year, R-squared: {model.rsquared}")

    axs[i, 0].set_title(f'Male vs Female Population in {borough}, 2010-2020', pad=20)
    axs[i, 0].set_xlabel('Year')
    axs[i, 0].set_ylabel('Population')
    axs[i, 0].legend(loc='upper left')

    # Calculate total male and female population percentage for the pie chart
    total_male_percentage = borough_data.groupby('Year')['Male Percentage'].mean() / 100
    total_female_percentage = 1 - total_male_percentage

    # Create a pie chart on the second subplot
    if year in total_male_percentage.index:
        male_percentage = total_male_percentage.loc[year]
        female_percentage = total_female_percentage.loc[year]
        axs[i, 1].pie([male_percentage, female_percentage], labels=['Male', 'Female'], colors=colors, autopct='%1.1f%%')
        axs[i, 1].set_title(f'Male vs Female Population Percentage in {borough}, {year}', pad=20)
        axs[i, 1].legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust the layout
plt.tight_layout()
plt.show()
```

- **Intercept (b)**: Represents the estimated population at the starting point of the analysis (where the time variable is zero). This might reflect the baseline population size before the observed period.
- **R Value**: Represents how well the data fits the model. A higher R value (close to 1) indicates a strong correlation between time and population size, suggesting that the model explains the variation in population size well.
